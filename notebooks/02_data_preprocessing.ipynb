```python
import pandas as pd

df = pd.read_csv("raw_used_cars.csv")

print("--- 1. DATASET SHAPE & DATA TYPES ---")
df.info()

print("\n--- 2. MISSING VALUES COUNT ---")
print(df.isnull().sum())

print("\n--- 3. NUMERICAL STATISTICS ---")
display(df.describe())

print("\n--- 4. CATEGORICAL UNIQUE VALUES ---")
categorical_cols = ['Condition', 'Transmission', 'Fuel type']
for col in categorical_cols:
    if col in df.columns:
        print(f"\nUnique values in '{col}':")
        print(df[col].value_counts(dropna=False))
```

```python
import numpy as np

initial_rows = len(df)

print("Starting data cleaning...")

if 'Condition' in df.columns:
    df = df.drop(columns=['Condition'])

# Handle Missing Values
df['Trim / Edition'] = df['Trim / Edition'].fillna("Unknown")
df['Body type'] = df['Body type'].fillna("Unknown")

# Filter mathematical outliers
df_clean = df[
    (df['Selling Price'] >= 500000) &
    (df['Selling Price'] <= 100000000) &
    (df['Engine capacity'] >= 600) &
    (df['Engine capacity'] <= 5000) &
    (df['Mileage'] >= 500) &
    (df['Mileage'] <= 500000) &
    (df['Year of Manufacture'] >= 1990)
]

# Clean up the Categorical Text
categorical_columns = ['Brand', 'Model', 'Trim / Edition', 'Transmission', 'Body type', 'Fuel type']
for col in categorical_columns:
    df_clean.loc[:, col] = df_clean[col].astype(str).str.strip().str.title()

print("\n--- CLEANING RESULTS ---")
print(f"Original Rows: {initial_rows}")
print(f"Cleaned Rows:  {len(df_clean)}")
print(f"Rows Removed:  {initial_rows - len(df_clean)}")

print("\n--- REMAINING MISSING VALUES ---")
print(df_clean.isnull().sum())

df_clean.to_csv("clean_used_cars.csv", index=False)
print("\nSuccess! Clean dataset saved as 'clean_used_cars.csv'")
```
